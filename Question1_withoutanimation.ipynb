```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse, Circle
import pprint
import numpy
from numpy import *
#Set axes aspect to equal as orbits are almost circular; hence square is needed
#ax = plt.figure(0).add_subplot(111, aspect='equal')


#Setting the title, axis labels, axis values and introducing a grid underlay
#Variable used so title can indicate user inputed date
plt.title('Inner Planetary Orbits')
plt.ylabel('x10^6 km')
plt.xlabel('x10^6 km')
#ax.set_xlim(-7,7)
#ax.set_ylim(-7,7)
plt.grid()

#Creating the point to represent the sun at the origin (not to scale),
plt.scatter(0,0,s=200,color='y')
plt.annotate('M', xy=(1,-3))

def thetaradius(R):
    m1= 3.28*10**23
    m2= 4*10**23
#     R= 2
    M = 1.99*10**30     #M = mass of sun
    a = R      #a = semi-major axis
    G = 6.674*10**-11   #G = Gravitationl constant
    k = G*M*m1        
    E = -k/(2*a)        #E = energy
    L= ((2*M*R**2)/5)*(3.984*10**-7) #L = angular momentum
    p = L**2/(m1*k)  
    e = 0        #e = eccentricity

    def fx(x):
        r = p/(1 + e*cos(x))
        return r

    n = 1000
    phi =linspace(0,2*pi,n)
    radius = zeros([n])
    theta = zeros([n])
    x = zeros([n])
    y = zeros([n])

    for i in range(0,n):
        radius[i] = fx(phi[i])
        theta[i] = 180*phi[i]/pi

    for i in range(0,n):
        x[i] = radius[i]*cos(phi[i])

    for i in range(0,n):
        y[i] = radius[i]*sin(phi[i])

   
    return x,y
a,b=thetaradius(2)
p,q=thetaradius(3)
plt.plot(a,b)
plt.plot(p,q)
plt.show()
```
